### 1. Bar Charts: Comparing Categories 📊

#### What Are Bar Charts?
Bar charts use bars to compare different categories. For example:
- Which hobby is most popular in your class?
- How much time do you spend on different activities?

Imagine you surveyed your classmates about their favorite snacks. Here’s the data:
- **Chips**: 15 votes
- **Chocolate**: 25 votes
- **Popcorn**: 10 votes
- **Candy**: 20 votes

Let’s visualize this with a bar chart!

````python
import matplotlib.pyplot as plt

# Favorite Snacks
data = {
    "Snacks": ["Chips", "Chocolate", "Popcorn", "Candy"],
    "Votes": [15, 25, 10, 20]
}

# Create a bar chart
plt.bar(data["Snacks"], data["Votes"], color="skyblue")
plt.title("Favorite Snacks")
plt.xlabel("Snacks")
plt.ylabel("Votes")
plt.show()
````

#### Advanced Example: Grouped Bar Charts
Let’s compare how boys and girls voted for these snacks. We’ll create a **grouped bar chart** to show both datasets side by side.

````python
# Grouped Bar Chart Example
import numpy as np

snacks = ["Chips", "Chocolate", "Popcorn", "Candy"]
boys_votes = [10, 12, 8, 14]
girls_votes = [5, 13, 2, 6]

x = np.arange(len(snacks))

plt.bar(x - 0.2, boys_votes, width=0.4, label="Boys", color="blue")
plt.bar(x + 0.2, girls_votes, width=0.4, label="Girls", color="pink")
plt.xticks(x, snacks)
plt.title("Snack Votes by Gender")
plt.xlabel("Snacks")
plt.ylabel("Votes")
plt.legend()
plt.show()
````

### 2. Line Graphs: Tracking Trends Over Time 📈

#### What Are Line Graphs?
Line graphs show how something changes over time. For example:
- How your energy levels vary during the day.
- How the number of books you read grows month by month.

Imagine you tracked your study hours for a week. Here’s the data:
- **Monday**: 2 hours
- **Tuesday**: 2.5 hours
- **Wednesday**: 3 hours
- **Thursday**: 4 hours
- **Friday**: 3.5 hours
- **Saturday**: 5 hours
- **Sunday**: 4.5 hours

Let’s plot this as a line graph!

````python
# Weekly Study Hours
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
study_hours = [2, 2.5, 3, 4, 3.5, 5, 4.5]

plt.plot(days, study_hours, marker="", linestyle="--", color="purple")
plt.title("Weekly Study Hours")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
plt.show()
````

### 3. Scatter Plots: Exploring Relationships 📍

#### What Are Scatter Plots?
Scatter plots help you find relationships between two things. For example:
- Does more studying lead to higher test scores?
- Does more sleep improve your energy levels?

Imagine you tracked your screen time and how much sleep you got. Let’s explore the relationship between screen time and sleep hours.

````python
# Screen Time vs Sleep Hours
screen_time = [1, 2, 3, 4, 5, 6]
sleep_hours = [9, 8.5, 8, 7.5, 7, 6.5]

plt.scatter(screen_time, sleep_hours, color="orange")
plt.title("Screen Time vs Sleep Hours")
plt.xlabel("Screen Time (hours)")
plt.ylabel("Sleep Hours")
plt.show()
````

### 4. Pie Charts: Showing Proportions 🍰

#### What Are Pie Charts?
Pie charts show how a whole is divided into parts. They’re perfect for proportions, like:
- How you spend your day (sleep, study, play).
- What percentage of students are in different clubs.

Imagine tracking your daily activities. Here’s how you spent your day:
- **Sleep**: 8 hours
- **Study**: 4 hours
- **Play**: 6 hours
- **Eat**: 6 hours

Let’s create a pie chart to show this breakdown!

````python
# Daily Activities Breakdown
activities = ["Sleep", "Study", "Play", "Eat"]
hours = [8, 4, 6, 6]

plt.pie(hours, labels=activities, autopct="%1.1f%%", startangle=90)
plt.title("Daily Activities")
plt.show()
````

### 5. Histograms: Understanding Data Distribution 📊

#### What Are Histograms?
Histograms show how data is distributed across ranges. For example:
- How many people scored in each grade range (A, B, C, etc.).
- How often you spend different amounts of time on your phone.

Imagine you tracked how many minutes you spent on your phone across 10 days:
- **Minutes**: [30, 45, 50, 60, 65, 70, 80, 85, 100, 120]

Let’s use a histogram to see the distribution!

````python
# Phone Usage Histogram
minutes = [30, 45, 50, 60, 65, 70, 80, 85, 100, 120]

plt.hist(minutes, bins=5, color="teal", edgecolor="black")
plt.title("Phone Usage Distribution")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
````

### 6. Heatmaps: Spotting Patterns 🔥

#### What Are Heatmaps?
Heatmaps are colorful grids that show patterns in data. They’re great for:
- Comparing test scores across subjects and students.
- Tracking activity levels over time.

Imagine tracking your energy levels throughout the day over a week. Let’s visualize it with a heatmap!

````python
import seaborn as sns
import numpy as np

# Energy Levels Heatmap
data = np.random.randint(1, 10, size=(7, 5))  # Random energy levels for 7 days, 5 time slots
sns.heatmap(data, annot=True, cmap="coolwarm", cbar=True)
plt.title("Energy Levels Over the Week")
plt.xlabel("Time Slots")
plt.ylabel("Days")
plt.show()
````
