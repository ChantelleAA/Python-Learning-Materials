## What is Machine Learning?

Machine Learning is teaching computers to learn patterns from data, just like humans learn from experience. It’s used in:
- Self-driving cars to recognize objects
- Voice assistants like Alexa and Siri to understand speech
- Medical AI to detect diseases early

#### Analogy: Teaching Computers Like Toddlers
Imagine teaching a toddler to recognize a cat. You show them many pictures of cats, and over time, they learn to identify one. Machine Learning works similarly—computers are shown data, and they learn patterns from it.
